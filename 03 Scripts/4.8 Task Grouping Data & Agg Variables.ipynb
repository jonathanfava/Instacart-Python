### Q6.

```python
# create column for avg spent by customer

ords_prods_merge['avg_spent'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# check output

ords_prods_merge.head()
```

```python
# create spending flag column

ords_prods_merge.loc[ords_prods_merge['avg_spent'] < 10, 'spending_flag'] = 'Low spender'
ords_prods_merge.loc[ords_prods_merge['avg_spent'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# check output

ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
# check output

ords_prods_merge.head()
```

### Q7.

```python
pd.options.display.max_columns = None
```

```python
# create column for median days since prior order 

ords_prods_merge['med_ord_freq'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# check output

ords_prods_merge.head()
```

```python
# create frequency flag column

ords_prods_merge.loc[ords_prods_merge['med_ord_freq'] > 20, 'frequency_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['med_ord_freq'] <= 20) & (ords_prods_merge['med_ord_freq'] > 10), 'frequency_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['med_ord_freq'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
# check output

ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```
