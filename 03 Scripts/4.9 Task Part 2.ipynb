### Q5.

```python
# create a list holding True/False values to the test np.random.rand() <= 0.7

np.random.seed(4)
dev = np.random.rand(len(opc)) <= 0.7
```

```python
# store 70% of the sample in the daf big

big = opc[dev]
```

```python
# store 30% of the sample in the df small

small = opc[~dev]
```

```python
# check df size

len(opc)
```

```python
# check size of new dfs big and small

len(big) + len(small)
```

```python
# create a df for the small sample with the two necessary columns

df_2 = small[['time_of_order','prices']]
```

```python
# create line chart

line = sns.lineplot(data = df_2, x = 'time_of_order', y = 'prices')
```

### Q6.

```python
# create a df for the small sample with the two necessary columns

df_3 = small[['age','#_dependants']]
```

### Q7. 

```python
# create scatterplot for age v. spending power

scat = sns.scatterplot(x = 'age', y = 'income',data = opc)
```

### The scatterplot shows the relationship between age of a customer and their spending power. We can conlcude that the number of customers earning <200k is consistent across all ages. But the number of customers earning >200k and <400k are mostly over the age of 40 and the customers earning >400k a year are all over the age of 40.
